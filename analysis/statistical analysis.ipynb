```python
#basic statistical analysis
#library import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
#length distribution of reviews in different rating level
yelp = pd.read_csv('data/webscraper.csv')
print 'good'
yelp['text length'] = yelp['content'].apply(len)
g = sns.FacetGrid(data=yelp, col='rating')
g.map(plt.hist, 'text length', bins=50)
plt.show()
```

```python
#the number of each rating
rating_count = yelp['rating'].value_counts()
print rating_count
```

```python
yelp.info()
yelp.describe()
```

```python
#text length in box graph
sns.boxplot(x='rating', y='text length', data=yelp)
plt.show()
```

```python
rating = yelp.groupby('rating').mean()
print rating
yelp.corr()
sns.heatmap(data=yelp.corr(), annot=True)
plt.show()
```

```python
#corelation level
yelp_w = yelp.loc[:,['rating','useful','funny','cool','checkin','text length']]
yelp_w.corr()
sns.heatmap(data=yelp_w.corr(), annot=True, center=0,vmax=1,vmin=-1)
plt.show() 
```
